```python
import os
import warnings
warnings.simplefilter('ignore')
```

```python
import numpy as np
import pandas as pd
```

```python
from skimage.io import imread,imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
mom=os.listdir("C:/mom")
```

```python
dad=os.listdir("C:/dad")
```

```python
she=os.listdir("C:/she")
```

```python
limit = 10

mom_images =[None]*limit
j=0
for i in mom:
    if(j<limit):
        mom_images[j] = imread("C:/mom/"+i)
        j+=1
    else:
        break
```

```python
limit = 10

dad_images =[None]*limit
j=0
for i in dad:
    if(j<limit):
        dad_images[j] = imread("C:/dad/"+i)
        j+=1
    else:
        break
```

```python
limit = 10

she_images =[None]*limit
j=0
for i in she:
    if(j<limit):
        she_images[j] = imread("C:/she/"+i)
        j+=1
    else:
        break
```

```python
imshow(mom_images[0])
```

```python
imshow(dad_images[0])
```

```python
imshow(she_images[0])
```

```python
mom_gray =[None]*limit
j=0

for i in mom:
    if(j<limit):
        mom_gray[j]=rgb2gray(mom_images[j])
        j+=1
    else:
        break
```

```python
dad_gray =[None]*limit
j=0

for i in dad:
    if(j<limit):
        dad_gray[j]=rgb2gray(dad_images[j])
        j+=1
    else:
        break
```

```python
she_gray =[None]*limit
j=0

for i in she:
    if(j<limit):
        she_gray[j]=rgb2gray(she_images[j])
        j+=1
    else:
        break
```

# Training

```python
family=pd.concat([mom_data,dad_data,she_data])
```

```python
family
```

```python
from sklearn.utils import shuffle
```

```python
family_indexed = shuffle(family).reset_index()
```

```python
family_indexed
```

```python
family_in=family_indexed.drop(['index'],axis=1)
```

```python
family_in
```

```python
x = family_in.values[:,:-1]
```

```python
y = family_in.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.37,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from matplotlib import pyplot as plt
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
